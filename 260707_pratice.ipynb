```python
a, b = 17, 5

print(f"{a} + {b}  = {a + b}")
print(f"{a} - {b}  = {a - b}")
print(f"{a} * {b}  = {a * b}")
print(f"{a} / {b}  = {a / b}")
print(f"{a} // {b}  = {a // b}")
print(f"{a} % {b}  = {a % b}")
print(f"2 ** 10 = {2**10}")
```

```python
#2번 문제
num1, num2 = map(int, input("두 수를 입력하세요: ").split())
print(f"두 수의 합: {num1+num2}")
print(f"두 수의 차: {num1-num2}")
print(f"두 수의 곱: {num1*num2}")
print(f"두 수의 나눗셈: {num1/num2}")
print(f"두 수의 나머지: {num1%num2}")
print(f"두 수의 거듭제곱: {num1**num2}")
```

```python
#3번 문제
sec = int(input("총 초를 입력하세요: "))
h = sec//3600
m = (sec%3600)//60
s = sec%60

print(f"{h}시간 {m}분 {s}초")
```
