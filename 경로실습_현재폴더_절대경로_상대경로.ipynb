## 1. 나는 지금 어디에 있는가 - 현재 작업 폴더 확인

주피터 노트북이 실행되는 순간, 파이썬은 특정 폴더를 기준으로 동작합니다.
이 폴더를 **현재 작업 폴더(Current Working Directory, 줄여서 CWD)** 라고 부릅니다.

파일을 열거나 저장할 때 폴더를 따로 지정하지 않으면, 무조건 이 폴더가 기준이 됩니다.
먼저 내 노트북이 어느 폴더에서 실행되고 있는지 확인해 봅시다.

```python
import os

# 현재 작업 폴더를 확인합니다
current = os.getcwd()
print("현재 작업 폴더:", current)
```

## 2. 절대 경로 vs 상대 경로

경로를 표현하는 방법은 두 가지가 있습니다.

| 구분 | 절대 경로 | 상대 경로 |
|------|-----------|-----------|
| 기준 | 드라이브 최상위(뿌리)부터 전체 주소 | 현재 작업 폴더 |
| 예시(윈도우) | `C:\Users\kim\lecture\data\a.txt` | `data\a.txt` |
| 비유 | "경기도 화성시 OO로 12, 3층 회의실" | "지금 있는 건물에서 3층 회의실" |

- 절대 경로는 어디서 실행해도 항상 같은 곳을 가리킵니다. 대신 길고, 다른 컴퓨터에서는 안 맞을 수 있습니다.
- 상대 경로는 짧고 이식성이 좋습니다. 대신 **현재 작업 폴더가 어디냐에 따라 가리키는 곳이 달라집니다.**

상대 경로에서 자주 쓰는 특수 기호 두 가지는 꼭 기억하세요.

- `.` : 현재 폴더
- `..` : 한 단계 위(상위) 폴더

```python
# 현재 폴더(.)와 상위 폴더(..)가 실제로 어디인지 절대 경로로 변환해 봅니다
print("현재 폴더( . ) :", os.path.abspath("."))
print("상위 폴더( .. ):", os.path.abspath(".."))
```

출력 결과를 보면 `.` 은 현재 작업 폴더 자신이고, `..` 은 그보다 한 단계 위 폴더라는 것을 알 수 있습니다.

`os.path.abspath()` 는 상대 경로를 절대 경로로 바꿔 주는 함수입니다.
"이 상대 경로가 실제로 어디를 가리키는 거지?" 하고 헷갈릴 때 매우 유용합니다.

## 3. 노트북에서 폴더 만들기

이제 현재 작업 폴더 안에 `practice` 라는 폴더를 코드로 만들어 보겠습니다.
마우스 오른쪽 클릭 - 새 폴더 만들기를 코드로 하는 것과 완전히 같은 일입니다.

```python
# practice 폴더를 만듭니다
# exist_ok=True : 이미 폴더가 있어도 에러를 내지 않고 넘어갑니다
os.makedirs("practice", exist_ok=True)

print("폴더 생성 완료")
#os.path.abspath 상대 경로를 현재 작업 폴더 기준의 절대 경로로 변환
print("생성된 위치:", os.path.abspath("practice"))
```

### 잠깐, 탐색기로 확인

여기서 중요한 점: 코드에는 `"practice"` 라고 폴더 이름만 적었는데,
폴더는 **현재 작업 폴더 안에** 만들어졌습니다.
경로를 생략하면 항상 현재 작업 폴더가 기준이 된다는 것, 이것이 상대 경로의 핵심입니다.

이번에는 폴더 안에 폴더, 즉 하위 폴더를 한 번에 만들어 봅시다.

```python
# practice 폴더 안에 data 폴더와 result 폴더를 만듭니다
# 경로 구분은 / 를 쓰면 윈도우, 맥, 리눅스 어디서나 동작합니다
os.makedirs("practice/data", exist_ok=True)
os.makedirs("practice/result", exist_ok=True)

print("하위 폴더 생성 완료")
```

## 4. 노트북에서 파일 만들기

이번에는 텍스트 파일을 만들어 보겠습니다.
`open()` 함수에 `"w"` (write, 쓰기) 모드를 주면 파일이 새로 만들어집니다.

```python
# 현재 작업 폴더에 memo.txt 파일을 만듭니다
with open("memo.txt", "w", encoding="utf-8") as f:
    f.write("첫 번째 줄: 경로 실습 중입니다.\n")
    f.write("두 번째 줄: 이 파일은 현재 작업 폴더에 만들어졌습니다.\n")

print("파일 생성 완료:", os.path.abspath("memo.txt"))
```

```python
# 이번에는 상대 경로를 이용해 practice/data 폴더 안에 파일을 만듭니다
with open("practice/data/sample.txt", "w", encoding="utf-8") as f:
    f.write("이 파일은 practice 폴더 안의 data 폴더에 있습니다.\n")

print("파일 생성 완료:", os.path.abspath("practice/data/sample.txt"))
```

## 6. 가장 흔한 에러 직접 만나 보기 - FileNotFoundError

일부러 틀린 경로로 파일을 열어 보겠습니다.
에러 메시지를 미리 읽어 두면, 나중에 실제로 만났을 때 당황하지 않습니다.

```python
# 존재하지 않는 경로로 파일을 열어 봅니다
try:
    with open("data/sample.txt", "r", encoding="utf-8") as f:
        print(f.read())
except FileNotFoundError as e:
    print("에러 발생:", e)
    print()
    print("원인 분석: 파이썬은 아래 위치에서 파일을 찾았습니다.")
    print(" ->", os.path.abspath("data/sample.txt"))
    print("하지만 실제 파일 위치는 여기입니다.")
    print(" ->", os.path.abspath("practice/data/sample.txt"))
```

파일은 분명히 존재하는데 왜 에러가 났을까요?

`sample.txt` 는 `practice/data` 안에 있는데, 코드에는 `data/sample.txt` 라고 적었기 때문입니다.
상대 경로는 현재 작업 폴더 기준이므로, 파이썬은 현재 폴더 바로 아래의 `data` 폴더를 찾았고, 그런 폴더는 없었던 것입니다.

**FileNotFoundError를 만나면 가장 먼저 할 일**

1. `os.getcwd()` 로 현재 작업 폴더 확인
2. `os.path.abspath("내가 쓴 경로")` 로 파이썬이 실제로 어디를 찾고 있는지 확인
3. 탐색기에서 실제 파일 위치와 비교

이 세 단계만 기억하면 경로 문제의 90%는 스스로 해결할 수 있습니다.

```python
# 올바른 상대 경로로 다시 읽어 봅니다
with open("practice/data/sample.txt", "r", encoding="utf-8") as f:
    print(f.read())
```

## 7. 작업 폴더가 바뀌면 상대 경로도 달라진다

`os.chdir()` 를 쓰면 현재 작업 폴더를 다른 곳으로 옮길 수 있습니다.
작업 폴더를 `practice` 로 옮긴 뒤, 같은 상대 경로가 어떻게 달라지는지 확인해 봅시다.

```python
# 이동 전 위치를 변수에 저장해 둡니다 (나중에 돌아오기 위해서)
start_dir = os.getcwd()
print("이동 전:", start_dir)

# practice 폴더로 작업 폴더를 이동합니다
os.chdir("practice")
print("이동 후:", os.getcwd())
```

```python
# 이제 현재 작업 폴더가 practice 이므로,
# data/sample.txt 라는 상대 경로가 정상적으로 동작합니다
with open("data/sample.txt", "r", encoding="utf-8") as f:
    print(f.read())

# 아까는 에러가 났던 경로인데, 지금은 잘 읽힙니다.
# 상대 경로의 의미가 '현재 작업 폴더 기준'으로 바뀌었기 때문입니다.
```

```python
# .. 을 이용해 상위 폴더로 돌아갑니다
os.chdir("..")
print("복귀 후:", os.getcwd())

# 저장해 둔 시작 위치와 같은지 확인합니다
print("시작 위치와 동일한가?", os.getcwd() == start_dir)
```

**정리하면**

- 같은 상대 경로라도 현재 작업 폴더가 다르면 다른 곳을 가리킵니다.
- 그래서 협업하거나 코드를 배포할 때는 절대 경로를 하드코딩하지 않고, 상대 경로 또는 코드로 계산한 경로를 사용하는 것이 일반적입니다.

참고: `os.chdir()` 는 편리하지만 노트북 중간에 남발하면 셀 실행 순서에 따라 결과가 달라져 혼란스러워집니다. 실무에서는 꼭 필요한 경우가 아니면 작업 폴더는 그대로 두고 경로를 조합해서 쓰는 편이 안전합니다.

## 8. 탐색기를 코드로 열기 (보너스)

매번 경로를 복사해서 붙여넣는 것이 번거롭다면, 코드로 탐색기를 바로 열 수도 있습니다.

```python
import platform
import subprocess

target = os.getcwd()
system = platform.system()

try:
    if system == "Windows":
        os.startfile(target)            # 윈도우: 탐색기 열기
    elif system == "Darwin":
        subprocess.run(["open", target])   # 맥: Finder 열기
    else:
        subprocess.run(["xdg-open", target])  # 리눅스
    print(f"{system} 환경에서 폴더 열기를 시도했습니다:", target)
except Exception as e:
    print("이 환경에서는 폴더를 열 수 없습니다:", e)
```

## 10. 실습 폴더 정리 (선택)

실습이 끝난 뒤 만들었던 폴더와 파일을 지우고 싶다면 아래 셀을 실행하세요.
`shutil.rmtree()` 는 폴더를 통째로 삭제하며, **휴지통을 거치지 않고 즉시 삭제되므로** 경로를 반드시 확인하고 실행해야 합니다.

계속 남겨 두고 복습하고 싶다면 실행하지 않아도 됩니다.

```python
import shutil

# 삭제할 대상 목록
targets = ["practice", "mywork", "memo.txt"]

for t in targets:
    if os.path.isdir(t):
        shutil.rmtree(t)
        print("폴더 삭제:", t)
    elif os.path.isfile(t):
        os.remove(t)
        print("파일 삭제:", t)
    else:
        print("이미 없음:", t)

print()
print("정리 후 현재 폴더 내용:", os.listdir("."))
```

## 오늘 배운 내용 요약

| 함수 / 기호 | 역할 |
|-------------|------|
| `os.getcwd()` | 현재 작업 폴더 확인 |
| `os.path.abspath(경로)` | 상대 경로를 절대 경로로 변환 |
| `os.makedirs(경로, exist_ok=True)` | 폴더 생성 (중간 폴더까지 한 번에) |
| `open(경로, "w")` / `open(경로, "r")` | 파일 쓰기 / 읽기 |
| `os.listdir(경로)` | 폴더 안의 목록 확인 |
| `os.path.exists(경로)` | 파일 또는 폴더 존재 여부 확인 |
| `os.chdir(경로)` | 현재 작업 폴더 이동 |
| `.` / `..` | 현재 폴더 / 상위 폴더 |
