```python
#XOR 연산, 은닉층 노드 4개로 수정

import numpy as np
import tensorflow as tf


X = np.array([[0,0], [0,1], [1,0], [1,1]], dtype=np.float32)
Y = np.array([[0], [1], [1], [0]])

seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)

#은닉층 가중치와 편향
W1 = tf.Variable(tf.random.normal([2, 4]), dtype=tf.float32)
b1 = tf.Variable(tf.random.normal([4]), dtype=tf.float32)
#출력층 가중치와 편향
W2 = tf.Variable(tf.random.normal([4, 1]), dtype=tf.float32)
b2 = tf.Variable(tf.random.normal([1]), dtype=tf.float32)

def hypothesis(X) :
    return 1 / (1+tf.exp(-X))

#다층 퍼셉트론 정의 (은닉층 -> 출력층)
def perceptron(X):
    hidden = hypothesis(tf.matmul(X, W1) + b1)
    return hypothesis(tf.matmul(hidden, W2)+b2)

def cost(y_true, y_pred):
    return tf.reduce_mean(-(y_true * tf.math.log(y_pred) 
                            + (1-y_true) * tf.math.log(1-y_pred)))

opt = tf.keras.optimizers.SGD(learning_rate=0.5)

def binary_output(X):
    return int(X.numpy()[0][0] > 0.5)

def accuracy(y_true, y_pred):
    correct_prediction = tf.equal(tf.cast(y_pred > 0.5, tf.float32), y_true)
    return tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

for i in range(2001):
    with tf.GradientTape() as tape:
        pred = perceptron(X)
        loss = cost(Y, pred)
    grads = tape.gradient(loss, [W1, W2 ,b1, b2])
    opt.apply_gradients(zip(grads, [W1, W2, b1, b2]))

    if i % 200 == 0:
        W1_val = np.squeeze(W1.numpy())
        b1_val = np.squeeze(b1.numpy())
        W2_val = np.squeeze(W2.numpy())
        b2_val = np.squeeze(b2.numpy())
        print(f"Step: {i}, Loss: {loss.numpy():.9f}")
        print(f"W1 = {W1_val}, b1 = {b1_val}")
        print(f"W2 = {W2_val}, b2 = {b2_val}")

print("\n--------예측결과------------")

for input_data in X:
    prediction = perceptron(np.array([input_data], dtype=np.float32))
    print(f"입력: {input_data.tolist()}, 예측(확률): {prediction.numpy()[0][0]:.4f}, 예측(이진): {binary_output(prediction)}")


predictions = perceptron(X)
acc = accuracy(Y, predictions)
print(f"정확도: {acc.numpy():.4f}")
```

```python
#XOR 연산, 은닉층을 2개 만들기

import numpy as np
import tensorflow as tf


X = np.array([[0,0], [0,1], [1,0], [1,1]], dtype=np.float32)
Y = np.array([[0], [1], [1], [0]])

seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)

#은닉층1 가중치와 편향
W1 = tf.Variable(tf.random.normal([2, 2]), dtype=tf.float32)
b1 = tf.Variable(tf.random.normal([2]), dtype=tf.float32)
#은닉층2 가중치와 편향
W2 = tf.Variable(tf.random.normal([2, 2]), dtype=tf.float32)
b2 = tf.Variable(tf.random.normal([2]), dtype=tf.float32)
#출력층 가중치와 편향
W3 = tf.Variable(tf.random.normal([2, 1]), dtype=tf.float32)
b3 = tf.Variable(tf.random.normal([1]), dtype=tf.float32)


def hypothesis(X) :
    return 1 / (1+tf.exp(-X))

#다층 퍼셉트론 정의 (은닉층1 -> 은닉층2 -> 출력층)
def perceptron(X):
    hidden1 = hypothesis(tf.matmul(X, W1) + b1)
    hidden2 = hypothesis(tf.matmul(hidden1, W2) + b2)
    return hypothesis(tf.matmul(hidden2, W3)+b3)

def cost(y_true, y_pred):
    return tf.reduce_mean(-(y_true * tf.math.log(y_pred) 
                            + (1-y_true) * tf.math.log(1-y_pred)))

opt = tf.keras.optimizers.SGD(learning_rate=0.5)

def binary_output(X):
    return int(X.numpy()[0][0] > 0.5)

def accuracy(y_true, y_pred):
    correct_prediction = tf.equal(tf.cast(y_pred > 0.5, tf.float32), y_true)
    return tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

for i in range(2000):
    with tf.GradientTape() as tape:
        pred = perceptron(X)
        loss = cost(Y, pred)
    grads = tape.gradient(loss, [W1, W2, W3, b1, b2, b3])
    opt.apply_gradients(zip(grads, [W1, W2, W3, b1, b2, b3]))

    if i % 200 == 0:
        W1_val = np.squeeze(W1.numpy())
        b1_val = np.squeeze(b1.numpy())
        W2_val = np.squeeze(W2.numpy())
        b2_val = np.squeeze(b2.numpy())
        W3_val = np.squeeze(W3.numpy())
        b3_val = np.squeeze(b3.numpy())
        print(f"Step: {i}, Loss: {loss.numpy():.9f}")
        print(f"W1 = {W1_val}, b1 = {b1_val}")
        print(f"W2 = {W2_val}, b2 = {b2_val}")
        print(f"W2 = {W3_val}, b2 = {b3_val}")

print("\n--------예측결과------------")

for input_data in X:
    prediction = perceptron(np.array([input_data], dtype=np.float32))
    print(f"입력: {input_data.tolist()}, 예측(확률): {prediction.numpy()[0][0]:.4f}, 예측(이진): {binary_output(prediction)}")


predictions = perceptron(X)
acc = accuracy(Y, predictions)
print(f"정확도: {acc.numpy():.4f}")
```

폐암 수술 환자의 생존율 예측
- Hidden layer 1개 (노드 30개)
- hidden layer와 hypothesis 모두 sigmoid

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt("Data/ThoraricSurgery.csv", delimiter=',')

x_train = data[:, 0:17].astype(np.float32)
y_train = data[:, [17]].astype(np.float32)

seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)

W1 = tf.Variable(tf.random.normal([17,30]), name = 'W1')
b1 = tf.Variable(tf.random.normal([30]), name = 'b1')
W2 = tf.Variable(tf.random.normal([30,1]), name = 'W2')
b2 = tf.Variable(tf.random.normal([1]), name = 'b2')


def forward(x):
    hidden = tf.sigmoid(tf.matmul(x, W1) + b1)
    hypothesis = tf.sigmoid(tf.matmul(hidden, W2) + b2)
    return hypothesis

def cost(y_true, y_pred):
    return tf.reduce_mean(-(y_true * tf.math.log(y_pred) 
                            + (1-y_true) * tf.math.log(1-y_pred)))

opt = tf.keras.optimizers.SGD(learning_rate=0.1)

def binary_output(X):
    return int(X.numpy()[0][0] > 0.5)

def accuracy(y_true, y_pred):
    correct_prediction = tf.equal(tf.cast(y_pred > 0.5, tf.float32), y_true)
    return tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

for i in range(2001):
    with tf.GradientTape() as tape:
        pred = forward(x_train)
        loss = cost(y_train, pred)
    grads = tape.gradient(loss, [W1, W2, b1, b2])
    opt.apply_gradients(zip(grads, [W1, W2, b1, b2]))

    if i % 200 == 0:
        print(f"Step: {i}, Loss: {loss.numpy():.9f}")
       

print("\n--------예측결과------------")

for input_data in x_train:
    prediction = forward(np.array([input_data], dtype=np.float32))
    print(f"입력: {input_data.tolist()}, 예측(확률): {prediction.numpy()[0][0]:.4f}, 예측(이진): {binary_output(prediction)}")


predictions = forward(x_train)
acc = accuracy(y_train, predictions)
print(f"정확도: {acc.numpy():.4f}")

```
