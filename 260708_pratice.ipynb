```python
target = 7

count = 0

while True:
    n = int(input("1부터 10 사이 숫자를 맞혀보세요: "))
    if target == n:
        count = count + 1
        print("정답입니다.")
        break

    elif target <= n <= 10:
        count = count + 1
        print("더 작은 숫자입니다.")

    elif target >= n >= 1:
        count = count + 1
        print("더 큰 숫자입니다.")

    else:
        print("다시 입력하시오.")

print(f"시도 횟수: {count}")
```
