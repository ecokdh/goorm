```python
import numpy as np

y_true = [21,49,68,80]
y_pred = [25.1,44.7,64.3,83.9]

def rsme(y_true, y_pred):
    mse = np.mean((np.array(y_true)- np.array(y_pred))**2)
    return np.sqrt(mse)

print("RSME: ", rsme(y_true, y_pred))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = [[2,81],[4,93],[6,91],[8,97]]
x = [i[0] for i in data]
y = [i[1] for i in data]

print(x,y)

plt.figure(figsize=(8,5))
plt.scatter(x,y)
plt.show()
```

```python
x_data = np.array(x)
y_data = np.array(y)

a = 0
b = 0

lr = 0.03
epochs = 2001

for i in range(epochs):
    a_diff = 2/len(x_data) * np.sum((a * x_data + b - y_data) * x_data)
    b_diff = 2/len(x_data) * np.sum(a*x_data + b - y_data)
    a = a - lr * a_diff
    b = b - lr * b_diff
    if i % 100 == 0:
        print("epoch=%.f, 기울기=%.04f, 절편=%.04f" % (i, a, b) )
```

```python
y_pred = a * x_data + b
plt.scatter(x,y)

plt.plot(x_data, y_pred)
plt.show()
```
