```python
import platform
import subprocess

target = os.getcwd()
system = platform.system()

try:
    if system == "Windows":
        os.startfile(target)            # мңҲлҸ„мҡ°: нғҗмғүкё° м—ҙкё°
    elif system == "Darwin":
        subprocess.run(["open", target])   # л§Ҙ: Finder м—ҙкё°
    else:
        subprocess.run(["xdg-open", target])  # лҰ¬лҲ…мҠӨ
    print(f"{system} нҷҳкІҪм—җм„ң нҸҙлҚ” м—ҙкё°лҘј мӢңлҸ„н–ҲмҠөлӢҲлӢӨ:", target)
except Exception as e:
    print("мқҙ нҷҳкІҪм—җм„ңлҠ” нҸҙлҚ”лҘј м—ҙ мҲҳ м—ҶмҠөлӢҲлӢӨ:", e)
```

```python
#л”Ҙлҹ¬лӢқмқ„ кө¬лҸҷн•ҳлҠ” лҚ° н•„мҡ”н•ң мјҖлқјмҠӨ н•ЁмҲҳ
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
#н•„мҡ”н•ң лқјмқҙлёҢлҹ¬лҰ¬
import numpy as np
import tensorflow as tf

#мӢӨн–үн•  л•Ңл§ҲлӢӨ к°ҷмқҖ кІ°кіјлҘј м¶ңл Ҙн•ҳкё° мң„н•ҙ м„Өм •н•ҳлҠ” мҪ”л“ң
np.random.seed(3)
tf.random.set_seed(3)
#мӨҖл№„лҸҲ мҲҳмҲ  нҷҳмһҗ лҚ°мқҙн„° л¶Ҳлҹ¬мҳӨкё°
Data_set = np.loadtxt("Data/ThoraricSurgery.csv", delimiter =",")
#нҷҳмһҗмқҳ кё°лЎқкіј мҲҳмҲ  кІ°кіјлҘј XмҷҖ YлЎң кө¬л¶„н•ҳм—¬ м ҖмһҘ
X = Data_set[:, 0:17]
Y = Data_set[:, 17]
#л”Ҙлҹ¬лӢқ кө¬мЎ°лҘј кІ°м • (лӘЁлҚёмқ„ м„Өм •н•ҳкі  мӢӨн–ү)
model = Sequential()
model.add(Input(shape = (17,))) #2.x нӣ„л°ҳм—җ ліҖкІҪлҗң л¶Җл¶„: Input л Ҳмқҙм–ҙ мӮ¬мҡ©
model.add(Dense(30, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
#л”Ҙлҹ¬лӢқ мӢӨн–ү
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.fit(X, Y, epochs = 100, batch_size = 10)
#кІ°кіј м¶ңл Ҙ
print("\n Accuracy: %.4f" % (model.evaluate(X, Y)[1]))

```
